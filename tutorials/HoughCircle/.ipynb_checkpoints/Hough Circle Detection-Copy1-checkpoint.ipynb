## Hough Circle Detection

Here is small tutorial on how to use hough circle detection on ancient objects.

The tutorial covers both the implementation of the algorithm and it's usage.


First a list of packages

```python
import numpy as np  # for matrix manipulation
from scipy import ndimage  # for convinient matrix operations on images
import imageio  # for image input output
import matplotlib.pyplot as plt  # plotting images
```

```python
# arranging plot sizes

fig_size = plt.rcParams["figure.figsize"] 
# Set figure width to 12 and height to 9
fig_size[0] = 21
fig_size[1] = 18
plt.rcParams["figure.figsize"] = fig_size
```

The following 3 code cells are boilerplate, that is they are necessary to make the algorithm work, but they don't belong to the algorithm.

We'll start with using a dummy image to demonstrate the algorithm

```python
# Let's read and plot the dummy image
im1 = imageio.imread('images/circlesimg.png')
plt.imshow(im1)
```

Here is a function that converts the image into grayscale

```python
def getGrayScale(img: np.ndarray([], dtype=np.uint8)):
    "Get grayscale image"
    # r: 0.2126
    # g: 0.7152
    # b: 0.0722
    img[:, :, 0] = img[:, :, 0] * 0.2126
    img[:, :, 1] = img[:, :, 1] * 0.7152
    img[:, :, 2] = img[:, :, 2] * 0.0722
    #
    grayim = img.sum(axis=2, dtype=np.uint8)
    #
    return grayim
```

```python
im1cp = im1.copy()
gray = getGrayScale(im1cp)
plt.imshow(gray, cmap='gray')
```

Let's make a function to find the edges of the image, since we only need the circumference in order to detect a circle

```python
def getEdges(grayim: np.ndarray([], dtype=np.uint8),
            ) -> np.ndarray([], dtype=np.int32):
    "Compute edges of the grayscale image"
    glimage = ndimage.laplace(grayim, 
                              #sigma=sigmas[1]
                             )
    #
    return glimage
```

```python
graycp = gray.copy()
edges = getEdges(graycp)
plt.imshow(edges, cmap='gray')
```

```python
def getSinCosTheta(img: np.ndarray([], dtype=np.uint8)):
    "Get sincos values of angles in 360 degree"
    rows, cols = img.shape[:2]
    sin = {}
    cos = {}
    thetas = np.linspace(0, 360, num=max(rows, cols), dtype=np.float32)
    #
    for angle in thetas:
        ang = angle * np.pi / 180
        sin[angle] = np.sin(ang)
        cos[angle] = np.cos(ang)
    #
    return sin, cos, thetas
```

```python
def voteAccumulator(img: np.ndarray([], dtype=np.uint8),
                    edgecoords: np.ndarray([], dtype=np.uint32),
                    radius_minmax=None,  # (int, int),
):
    #
    sins, coss, thetas = getSinCosTheta(img)

    rows = img.shape[0]
    cols = img.shape[1]

    if radius_minmax is not None:
        rmin = radius_minmax[0]
        rmax = radius_minmax[1]
    else:
        rmin = 0
        diagonalSqr = rows ** 2 + cols ** 2
        diagonal = np.sqrt(diagonalSqr)
        diagonal = np.int32(diagonal)
        rmax = diagonal // 2
    #
    accu = np.zeros((rows,
                     cols,
                     rmax), dtype=np.int32)
    #
    for x, y in edgecoords:
        for radius in range(rmin, rmax):
            for angle in thetas:                
                a = np.int(x - round(radius * sins[angle], 2))
                b = np.int(y - round(radius * coss[angle], 2))
                if (
                    (
                        (a >= 0 and  a < rows) and
                        (b >= 0 and b < cols)
                    )
                ):
                    accu[a, b, radius] += 1
    #
    return accu
```

There is a lot of computation that is happening in the next cell, so depending on your machine it could take a while.

```python
ecopy = edges.copy()
ecoords = np.argwhere(ecopy)
graycp = gray.copy()
accumulator = voteAccumulator(img=graycp, 
                              edgecoords=ecoords)
```
